```python

import h2o
from h2o.estimators import H2OPrincipalComponentAnalysisEstimator
h2o.init(min_mem_size='12G')
```

```python
mono_resistance_frame = h2o.import_file( "../data/processed/mono_resistance_frame.csv")
mono_resistance_frame.head()
```

```python
mono_resistance_frame.shape
```

```python
training, validation = mono_resistance_frame.split_frame(ratios = [.7], seed = 1234)
```

```python
# Build and train the model:
mono_resistance_frame_pca = H2OPrincipalComponentAnalysisEstimator(k = 1,
                                                                   pca_method = "GramSVD")

mono_resistance_frame_pca.train(training_frame = training)
```

```python
# Generate predictions on a validation set (if necessary):
pred = mono_resistance_frame_pca.predict(validation)
pred
```
