```python
# Find the relationship between drtype and main_lin variables
grouped = df.groupby(['drtype','main_lin'])
grouped.size()
```

```python
## Find the relationship between samples and drtype variables

drtype_sample_table = pd.crosstab(index=df.index, columns=df["drtype"])
drtype_sample_table
```

```python
## Find the relationship between main lineages and drtype variables

drtype_mainlin_table = pd.crosstab(index=df['main_lin'], columns=df["drtype"])
drtype_mainlin_table
```

```python
# Stacked Column Chart: visual form of the two-way table

drtype_mainlin_table.plot(kind="bar", 
                 figsize=(8,8),
                 stacked=True)
```

```python
## Find the relationship between sub lineages and drtype variables

drtype_sublin_table = pd.crosstab(index=df['sublin'], columns=df["drtype"])
drtype_sublin_table
```

```python
# Stacked Column Chart: visual form of the two-way table

drtype_sublin_table.plot(kind="bar", 
                 figsize=(8,8),
                 stacked=True)
```
